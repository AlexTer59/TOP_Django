```python
import requests
from bs4 import BeautifulSoup as BS
import re
articles_data = []
```

```python
for page in range(1, 11):
    text = requests.get(f'https://www.e1.ru/text/?page={page}').text
    soup = BS(text)
    articles = soup.find_all('article', {'class': 'OPHIx'})
    for article in articles:
        title = article.find('h2', {'class': 'h9Jmx'}).find('a').text
        try:
            views = article.find('span', {'class': '_3mETe'}).text
            views = int(''.join(re.findall(r'\d+', views)))
        except AttributeError:
            views = None
        articles_data.append({'title': title, 'views': views})
```

```python
import pandas as pd

df = pd.DataFrame(articles_data)
df.sort_values('views', ascending=False)
```
